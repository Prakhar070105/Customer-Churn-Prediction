# Dataset Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries Imported Successfully")
```

```python
sns.set_style('whitegrid')
%matplotlib inline
```

```python
# Loading dataset 
```

```python
df = pd.read_csv('../dataset/WA_Fn-UseC_-Telco-Customer-Churn.csv')
print("Data Loaded Successfully")
```

```python
# Shape of the Dataset

df.shape
```

```python
# Checking the Data Types and Missing Values in the dataset

df.info()
```

```python
# Total charge should be numeric but here it is in object - type conversion

df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df.info()
```

```python
# Finding Missing values - NaN/Null

df['TotalCharges'].isnull().sum()
```

```python
# Since 11 values are null, we will remove it 

df = df.dropna()
```

```python
df.info()
```

```python
# Saveing the cleaned dataframe to a new file in the dataset folder

df.to_csv('../dataset/cleaned_telco_churn.csv', index=False)
print("new dataset saved")
```

# 2. Data Vizualization

```python
print(df['Churn'].value_counts())
```

```python
# Bar Chart Visualization

plt.figure(figsize=(6, 4))
sns.countplot(x='Churn', data=df)
plt.title('Distribution of Customer Churn')
plt.xlabel('Churn')
plt.ylabel('Number of Customers')
plt.show()
```

```python
# Checking the Churn based on gender

plt.figure(figsize=(6, 4)) 
sns.countplot(x='gender', hue='Churn', data=df, palette='viridis', width=0.7)
plt.title('Churn by Gender', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Number of Customers', fontsize=12)
plt.legend(title='Churn', fontsize=10)
plt.show()
```

```python
# Analyzing and Vizualizing how Churn is impacted by Senior Citizens, Married person and Dependent people(children)

columns = ['SeniorCitizen', 'Partner', 'Dependents']

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for col, ax in zip(columns, axes): 
    sns.countplot(x=col, hue='Churn', data=df, ax=ax, palette='mako', width=0.6) 
    ax.set_title(f'Churn by {col}')
    ax.set_xlabel(col)
    ax.set_ylabel('Number of Customers')

plt.tight_layout()
plt.show()
```

```python
# Visualizing account-related columns to analyze

account_cols = ['Contract', 'PaperlessBilling', 'PaymentMethod']

fig, axes = plt.subplots(3, 1, figsize=(8, 12))

for col, ax in zip(account_cols, axes):
    sns.countplot(x=col, hue='Churn', data=df, ax=ax, palette='flare', width=0.6)
    ax.set_title(f'Churn by {col}', fontsize=14)
    ax.set_xlabel(col, fontsize=12)
    ax.set_ylabel('Number of Customers', fontsize=12)

plt.tight_layout()  
plt.show()
```

```python
# Visualization of service-related columns to analyze

service_cols = ['InternetService', 'OnlineSecurity', 'TechSupport']

fig, axes = plt.subplots(3, 1, figsize=(8, 12))

for col, ax in zip(service_cols, axes):
    sns.countplot(x=col, hue='Churn', data=df, ax=ax, palette='pastel', width=0.6) 
    ax.set_title(f'Churn by {col}', fontsize=14)
    ax.set_xlabel(col, fontsize=12)
    ax.set_ylabel('Number of Customers', fontsize=12)

plt.tight_layout() 
plt.show()
```

```python
# Visualizing numerical columns with histogram 

numerical_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for col, ax in zip(numerical_cols, axes):
    sns.histplot(df[col], kde=True, ax=ax, color='green') 
    ax.set_title(f'Distribution of {col}', fontsize=14)
    ax.set_xlabel(col, fontsize=12)
    ax.set_ylabel('Density', fontsize=12) 

plt.tight_layout()
plt.show()
```

```python

numerical_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']
\
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
my_palette = {'No': '#007bff', 'Yes': '#dc3545'} 

for col, ax in zip(numerical_cols, axes):

    sns.kdeplot(data=df, x=col, hue='Churn', 
                fill=True, common_norm=False, 
                palette=my_palette, ax=ax)
    
    if col == 'tenure':
        ax.set_xlabel('Tenure (Months)', fontsize=12)
    elif col == 'MonthlyCharges':
        ax.set_xlabel('Monthly Charges ($)', fontsize=12)
    else:
        ax.set_xlabel('Total Charges ($)', fontsize=12)
        
    ax.set_title(f'Distribution of {col} by Churn', fontsize=14)
    ax.set_ylabel('Density', fontsize=12)

plt.tight_layout()
plt.show()
```
